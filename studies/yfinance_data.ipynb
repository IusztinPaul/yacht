```python
import yfinance
import mplfinance as mpf
```

# Define Constants

```python
TICKERS = ['AAPL']
START = '2020-01-01'
END = None
INTERVAL = '1d'
```

# Download Data

```python
tickers_data = yfinance.download(
            tickers=TICKERS,
            start=START,
            end=END,
            interval=INTERVAL,
            prepost=False
        )
```

```python
tickers_data.head()
```

```python
tickers_data = yfinance.download(
            tickers=TICKERS,
            start=START,
            end=END,
            interval=INTERVAL,
            auto_adjust=True,
            prepost=False
        )
```

```python
print(len(tickers_data))
tickers_data.head()
```

```python
tickers_data = yfinance.download(
            tickers=TICKERS,
            start=START,
            end=END,
            interval=INTERVAL,
            auto_adjust=True,
            prepost=False,
            actions=True
        )
```

```python
print(len(tickers_data))
tickers_data.head(n=40)
```

# Data Continuity

```python
mpf.plot(tickers_data.loc['2020-01-01':'2020-02-01'], type='candle')
```

We can see that there is noise between weekends or other holidays when the market is not open.

# Stock Splits Continuity

A stock split was made on the APPL stock on 31 August 2021. Lets see if the data is readjusted to the split.

```python
tickers_data.loc['2020-08-20':'2020-9-10']
```

```python
mpf.plot(tickers_data.loc['2020-07-01':'2020-10-01'], type='candle')
```

We can see that prices are adjusted to stock splits.
