```python
import yfinance
import mplfinance as mpf
```

```python
data = yfinance.download(
            tickers=['NVDA'],
            start='2021-01-01',
            end=None,
            interval='1d',
            auto_adjust=True,
            prepost=False
        )
```

```python
year_data = data['2021-01-01': '2021-02-01']
mean = year_data.mean()[['Open', 'High', 'Low', 'Close']].mean()
mpf.plot(year_data, volume=True, type='candle', hlines=dict(hlines=[mean],colors=['b'],linestyle='-.'))
```
